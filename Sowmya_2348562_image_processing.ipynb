normalization using min max

```python
img = cv2.imread('/content/normal.jpg')
```

```python
# Apply threshold to create a binary image
ret, thresh = cv2.threshold(img, 140, 255, cv2.THRESH_BINARY)
img_float = thresh.astype('float32')

# Normalize the binary image to the range [0, 1]
img_normalized = (img_float - img_float.min()) / (img_float.max() - img_float.min())

normalized_img = (img_normalized * 255).astype('uint8')

cv2_imshow(img)
# Visualize the normalized image
cv2_imshow(normalized_img)
```

rotation , zooming and flipping : data augmentation

```python
image =  cv2.imread('/content/cv1.jpeg')
```

```python
image =  cv2.imread('/content/cont.jpeg')
```

```python
# 1. Rotation
angle = 45

height, width = image.shape[:2]
center = (width // 2, height // 2)

rotation_matrix = cv2.getRotationMatrix2D(center, angle, 1.0)
rotated_image = cv2.warpAffine(image, rotation_matrix, (width, height))

# 2. Zooming
scale = 1.5

zoomed_image = cv2.resize(image, None, fx=scale, fy=scale, interpolation=cv2.INTER_LINEAR)

# 3. Flipping
flipped_image_horizontal = cv2.flip(image, 1)  # 1 for horizontal flip
flipped_image_vertical = cv2.flip(image, 0)    # 0 for vertical flip
```

```python
# Display original and augmented images
plt.figure(figsize=(10, 6))

plt.subplot(2, 3, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original')

plt.subplot(2, 3, 2)
plt.title('rotated')
plt.imshow(rotated_image)
```

```python
# Display original and augmented images
plt.figure(figsize=(10, 6))

plt.subplot(2, 3, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original')

plt.subplot(2, 3, 2)
plt.imshow(cv2.cvtColor(rotated_image, cv2.COLOR_BGR2RGB))
plt.title('Rotated')

plt.subplot(2, 3, 3)
plt.imshow(cv2.cvtColor(zoomed_image, cv2.COLOR_BGR2RGB))
plt.title('Zoomed')

plt.subplot(2, 3, 4)
plt.imshow(cv2.cvtColor(flipped_image_horizontal, cv2.COLOR_BGR2RGB))
plt.title('Horizontal Flip')

plt.subplot(2, 3, 5)
plt.imshow(cv2.cvtColor(flipped_image_vertical, cv2.COLOR_BGR2RGB))
plt.title('Vertical Flip')

plt.tight_layout()
plt.show()
```
